# Oil Investments

```python
df_oil=df['Oil Investments']
df_oil=pd.DataFrame(df_oil)
df_oil
```

```python
df_oil.plot()
```

```python
#Ho: It is non stationary
#H1: It is stationary

from statsmodels.tsa.stattools import adfuller
def adfuller_test(sales):
    result=adfuller(sales) # same as above ,variable name changed
    labels = ['ADF Test Statistic','p-value','#Lags Used','Number of Observations Used']
    for value,label in zip(result,labels):
        print(label+' : '+str(value) )
    if result[1] <= 0.05:
        print("strong evidence against the null hypothesis(Ho), reject the null hypothesis. Data  is stationary")
    else:
        print("weak evidence against null hypothesis, time series is non-stationary ")
    
```

```python
adfuller_test(df_oil['Oil Investments'])
```

```python
print(df_oil.shape)
train = df_oil['2009-12-31':'2014-12-31']
test = df_oil['2015-01-01':]
print(train.shape,test.shape)
```

```python
train.head(20)
```

```python
train['Sales First Difference']=train['Oil Investments']-train['Oil Investments'].shift(1)
```

```python
## Again test dickey fuller test
adfuller_test(train['Sales First Difference'].dropna())
```

```python
train['Sales First Difference'].plot()
```

```python
from pandas.plotting import autocorrelation_plot
autocorrelation_plot(train['Oil Investments'])
plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
import statsmodels.api as sm
```

```python
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(211)
# First 12 values are null so we will start from 13
fig = sm.graphics.tsa.plot_acf(train['Sales First Difference'].iloc[2:],lags=40,ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(train['Sales First Difference'].iloc[2:],lags=40,ax=ax2)
```

```python
from statsmodels.tsa.arima_model import ARIMA  
model=ARIMA(train['Oil Investments'],order=(1, 1, 1)) #(p,d,q,Shift)
results=model.fit()
print(results.summary())
train['forecast']=results.predict(start=25,end=1258,dynamic=True)
train[['Oil Investments','forecast']].plot(figsize=(12,8))
```

```python
print(train['forecast'][25:])
train['forecast'].plot()
```
