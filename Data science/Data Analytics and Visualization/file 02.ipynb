# Data Analytics and Visualization

```python
# Import pandas and numpy
import pandas as pd
import numpy as np 

# import visualization libraries :
import matplotlib.pyplot as plt
import seaborn as sns 

# Reading dataset
transfusion = pd.read_csv("transfusion.data")

# Print out the first rows of our dataset
transfusion.head()
```

```python
# SWARM PLOT

sns.set(style='whitegrid')
sns.swarmplot(y="Recency (months)",
                  x="Frequency (times)",
                  data=transfusion)

```

```python
cols=["Recency (months)","Frequency (times)",'Monetary (c.c. blood)','Time (months)']
```

```python
#BOX PLOT
for i in cols:
    sns.boxplot(y=transfusion[i], x=transfusion["whether he/she donated blood in March 2007"])
    plt.show()
```

```python
#Violin Plots
for i in cols:
    sns.violinplot(y=transfusion[i], x=transfusion["whether he/she donated blood in March 2007"])
    plt.show()
```

```python
#Line Plot
sns.lineplot(x="Monetary (c.c. blood)", y="Recency (months)", data=transfusion)
```

```python
#Scatter Plots
sns.scatterplot(x="Monetary (c.c. blood)", y="Recency (months)", data=transfusion)
```

```python
#Scatter Plots
sns.scatterplot(x="Monetary (c.c. blood)", y="Frequency (times)", data=transfusion)
```

```python
#Pair Plots

sns.pairplot(data=transfusion,hue="whether he/she donated blood in March 2007")
```

```python
#Histograms
plt.figure(figsize=(10,10))
transfusion.hist()
plt.plot()
plt.show()
```

```python
#Heat Maps
sns.heatmap(transfusion.corr())
```

```python
#Bar Plots
for i in cols:
    sns.barplot(y=transfusion[i], x=transfusion["whether he/she donated blood in March 2007"])
    plt.show()
```

```python
#Strip Plots

for i in cols:
    sns.stripplot(y=transfusion[i], x=transfusion["whether he/she donated blood in March 2007"])
    plt.show()
```

```python
#wCOUNTPLOT
sns.countplot(transfusion["whether he/she donated blood in March 2007"])
```

```python
#Factor Plots 
for i in cols:
    sns.factorplot(x=i, y="whether he/she donated blood in March 2007",data=transfusion)
    
```
